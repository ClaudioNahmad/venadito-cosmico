```python
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

#******************************************************

rc('font',**{'family':'serif','serif':['Times']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#*******************************************************
%matplotlib notebook
```

```python
# w ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

#N=1 **************************************************************************



#params3 b0=-0.9, b1=-0.2
n1pk3=np.loadtxt('./../nealika_N1/output/wn1/N1_params3_pk.dat')
n1cl3=np.loadtxt('./../nealika_N1/output/wn1/N1_params3_cl.dat')
n1len3=np.loadtxt('./../nealika_N1/output/wn1/N1_params3_cl_lensed.dat')

#N=2 **************************************************************************
#params1 b0=-0.1, b1=0.2, b2=-1
n2pk=np.loadtxt('./../piwami_N2/output/wn2/N2_pk.dat')
n2cl=np.loadtxt('./../piwami_N2/output/wn2/N2_cl.dat')
n2len=np.loadtxt('./../piwami_N2/output/wn2/N2_cl_lensed.dat')

#params2 b0=-0.85, b1=0.67, b2=-1
n2pk2=np.loadtxt('./../piwami_N2/output/wn2/N2_params2_pk.dat')
n2cl2=np.loadtxt('./../piwami_N2/output/wn2/N2_params2_cl.dat')

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

#LCDM **************************************************************************
lpk=np.loadtxt('./../../2_class_public/output/test_pk.dat')
lcl=np.loadtxt('./../../2_class_public/output/test_cl.dat')
```

```python
fig, ax = plt.subplots(figsize=(9,7))

#--------------------------------------------------------------------------------------
# N=1
ax.loglog(n1pk3[:,0],n1pk3[:,1],label='N=1, $b_0=-0.9, b_1=-0.2$',alpha=0.7,linewidth=1)

# N=2
#params1
ax.loglog(n2pk[:,0],n2pk[:,1],label='N=2, $b_0=-0.1, b_1=0.2, b_2=-1$',alpha=0.7,linewidth=1)

#params2
ax.loglog(n2pk2[:,0],n2pk2[:,1],label='N=2 $b_0=-0.85, b_1=0.67, b_2=-1$',alpha=0.7,linewidth=1)


# LCDM
ax.loglog(lpk[:,0],lpk[:,1],label=r'$\Lambda$-CDM',alpha=0.7,linewidth=1.5)
#--------------------------------------------------------------------------------------
ax.legend(loc='best', fancybox=True, framealpha=1)
ax.set_ylabel(r'$P(k) (Mpc/h)^3$',fontsize=15)
ax.set_xlabel(r'$k (h/Mpc) $',fontsize=15)
#ax.set_ylim(bottom=10**-5)
ax.set_xlim(left=2*10**-5,right=1)
ax.grid(which='major',ls=":", c='black',alpha=0.4)
#ax.set_title(r'CPL vs w(N=1) for parameters 1');
```
