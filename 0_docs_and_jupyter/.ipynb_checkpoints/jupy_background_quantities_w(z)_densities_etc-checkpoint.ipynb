```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

#******************************************************

rc('font',**{'family':'serif','serif':['Times']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#*******************************************************
%matplotlib notebook
```

```python
bkgrnd=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_12-1_f1.7_bf_background.dat')
bkgrnd_lcdm=np.loadtxt('../../class-public/output/lcdm_background.dat')
```

```python
bkgrnd[:,0]
```

```python
fig, ax = plt.subplots(figsize=(9,7))
ax.plot(bkgrnd[:,0],bkgrnd[:,3],'g',label=r'$w(N=1)$',alpha=0.5,linewidth=5)
ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,3],'k',label=r'$\Lambda - CDM$',alpha=1,linewidth=1)

#--------------------------------------------------------------------------------------
ax.set_xscale("log")
ax.set_yscale("log")

#ax.set_title(r'Planck 2018 TT Power Spectrum')
#ax.set_xlim(left=2, right=3000)
ax.set_xlabel(r'redshift $z$',fontsize=15)
ax.set_ylabel(r'$H(z)$',fontsize=15)
ax.legend(loc='upper right', fancybox=True, framealpha=1)
ax.grid(which='both',ls=":", c='black', alpha=0.4);
#--------------------------------------------------------------------------------------
# CAMBIAR NOMBRE AQUIIIIIIIIIIIIIII
#plt.savefig('tt_N1_lcdm.pdf')
```

```python
fig, ax = plt.subplots(figsize=(9,7))

ax.plot(bkgrnd[:,0],bkgrnd[:,8],label=r'$\rho_{\gamma}$',alpha=1,linewidth=1)
ax.plot(bkgrnd[:,0],bkgrnd[:,9],label=r'$\rho_{b}$',alpha=1,linewidth=1)
ax.plot(bkgrnd[:,0],bkgrnd[:,10],label=r'$\rho_{cdm}$',alpha=1,linewidth=1)
#ax.plot(bkgrnd[:,0],bkgrnd[:,11],label=r'$\rho_{\Lambda}$',alpha=1,linewidth=1)
ax.plot(bkgrnd[:,0],bkgrnd[:,12],label=r'$\rho_{fld}$',alpha=1,linewidth=1)
ax.plot(bkgrnd[:,0],bkgrnd[:,14],label=r'$\rho_{\nu}$',alpha=1,linewidth=1)

ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,8],label=r'$\rho_{\gamma}$2',alpha=1,linewidth=1)
ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,9],label=r'$\rho_{b}$2',alpha=1,linewidth=1)
ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,10],label=r'$\rho_{cdm}$2',alpha=1,linewidth=1)
ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,11],label=r'$\rho_{\Lambda}$',alpha=1,linewidth=1)
ax.plot(bkgrnd_lcdm[:,0],bkgrnd_lcdm[:,12],label=r'$\rho_{\nu}$2',alpha=1,linewidth=1)


#--------------------------------------------------------------------------------------
ax.set_xscale("log")
ax.set_yscale("log")

ax.set_title(r'$w(N=1)$ vs $\Lambda - CDM$ densities')
#ax.set_xlim(left=2, right=3000)
ax.set_xlabel(r'redshift $z$',fontsize=15)
ax.set_ylabel(r'density $\frac{8\pi G}{3} \rho$',fontsize=15)
ax.legend(loc='upper right', fancybox=True, framealpha=1)
#ax.grid(which='both',ls=":", c='black', alpha=0.4);
#--------------------------------------------------------------------------------------
# CAMBIAR NOMBRE AQUIIIIIIIIIIIIIII
plt.savefig('plot_wN1_lcdm_densities.pdf')
```
