```python
TThighl_2015=[]
for i in dataTThighl:
    TThighl_2015.append(i)
    
TTlowl_2015=[]
for j in dataTTlowl:
    TTlowl_2015.append(j)
```

```python
TThl=[0,0,0]
vector=[]


for i in TThighl_2015:
    ELL=i[0]
    D_ELL=i[1]
    ERR=i[2]
    vector.append(ELL)
    vector.append(D_ELL)
    vector.append(ERR)
    TThl=np.vstack((TThl,vector))
    vector=[]
```

```python
TTll=[0,0,0]
vector=[]


for j in TTlowl_2015:
    ELL=j[0]
    D_ELL=j[1]
    ERR=j[2]
    vector.append(ELL)
    vector.append(D_ELL)
    vector.append(ERR)
    TTll=np.vstack((TTll,vector))
    vector=[]
```

```python
TThl
```

```python
TTll
```

```python
TThighl_lowl=np.vstack((TTll,TThl))
TThighl_lowl
```

```python
x=TThighl_lowl[:,0]
y=TThighl_lowl[:,1]
err=TThighl_lowl[:,2]

fig, ax = plt.subplots(figsize=(9,7))
ax.errorbar(x,y,yerr=err,fmt='r.',ecolor='k',elinewidth=0.5,alpha=1,errorevery=4,
            markersize=0.7,label=r'Planck 2015')

#--------------------------------------------------------------------------------------
ax.set_xscale("log")
#ax.set_yscale("log")

ax.set_title(r'Planck 2015 TT Power Spectrum (Normal Plot)')
#ax.set_xlim(left=2, right=3000)
ax.set_xlabel(r'multipole moment $\ell$',fontsize=15)
ax.set_ylabel(r'$D_{\ell}^{TT} $',fontsize=15)
ax.legend(loc='upper right', fancybox=True, framealpha=1)
ax.grid(which='both',ls=":", c='black', alpha=0.4);
#--------------------------------------------------------------------------------------

#plt.savefig('planck_2015_Dl.pdf')
```
