```python
sample1 = getdist.loadMCSamples(r'/home/claudio/physics/Servicio-Social2/CosmoMC-RESULTS/lcdm_cmb/test_lcdm_cmb')
params=[u'omegabh2',u'omegach2',u'ns',u'H0']

g = plots.getSubplotPlotter()
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.4
g.triangle_plot(sample1, params, 
    filled_compare=True, 
    legend_labels=['Simulation'], 
    legend_loc='upper right', 
    line_args=[{'ls':'--', 'color':'green'}], 
    contour_colors=['green'])
```
