```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

#******************************************************

rc('font',**{'family':'serif','serif':['Times']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#*******************************************************
%matplotlib notebook
```

```python
#N=1 Nierika
pk1=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_12-1_f1.7_bf_pk.dat')
cl1=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_12-1_f1.7_bf_cl.dat')

#N=2 Nakawe
pk2=np.loadtxt('../N2_nakawe/output/nakawe_result/nakawe_1-3_f1.7_bf_pk.dat')
cl2=np.loadtxt('../N2_nakawe/output/nakawe_result/nakawe_1-3_f1.7_bf_cl.dat')

#LCDM **************************************************************************
lpk=np.loadtxt('../N1_nearika/output/lcdm/lcdm_planck2015_bf_pk.dat')
lcl=np.loadtxt('../N1_nearika/output/lcdm/lcdm_planck2015_bf_cl.dat')
```

```python
fig, ax = plt.subplots(figsize=(9,7))

#N=1 vs lcdm----------------------------------------------------------------------------
ax.loglog(pk1[:,0],pk1[:,1],'--y',label=r'$w(N=1)$',alpha=1,linewidth=1)
ax.loglog(pk2[:,0],pk2[:,1],'-g',label=r'$w(N=2)$',alpha=1,linewidth=1)
ax.loglog(lpk[:,0],lpk[:,1],'k',label=r'$\Lambda_{CDM}$',alpha=1,linewidth=1)

#ax.loglog(pk2[:,0],pk2[:,1],'--g',label=r'$w(N=1)2$',alpha=1,linewidth=1)
#ax.loglog(lpk2[:,0],lpk2[:,1],'b',label=r'$\Lambda_{CDM}2$',alpha=1,linewidth=1)




#--------------------------------------------------------------------------------------
ax.set_title(r'Planck $\Lambda_{CDM}$ vs BAO+JLA+Planck $w(N=1)$ and $w(N=2)$')
ax.legend(loc='best', fancybox=True, framealpha=1)
ax.set_ylabel(r'$P(k) (Mpc/h)^3$',fontsize=15)
ax.set_xlabel(r'$k (h/Mpc) $',fontsize=15)
#ax.set_ylim(bottom=10**-5)
ax.set_xlim(left=2*10**-5,right=1)
ax.grid(which='major',ls=":", c='black',alpha=0.4)
#ax.set_title(r'CPL vs w(N=1) for parameters 1');
#--------------------------------------------------------------------------------------


# CHANGE NAME HERE!!!!
plt.savefig('power_N1_vs_lcdm.pdf')
```
