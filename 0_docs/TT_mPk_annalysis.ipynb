```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

#******************************************************

rc('font',**{'family':'serif','serif':['Times']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#*******************************************************
%matplotlib notebook
```

```python
pk1=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_result1_pk.dat')
cl1=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_result1_cl.dat')

#LCDM **************************************************************************
lpk=np.loadtxt('../N1_nearika/output/lcdm/lcdm_planck2015pk.dat')
lcl=np.loadtxt('../N1_nearika/output/lcdm/lcdm_planck2015cl.dat')


pk2=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_2_pk.dat')
cl2=np.loadtxt('../N1_nearika/output/nierika_result_1/nierika_2_cl.dat')

lpk2=np.loadtxt('../N1_nearika/output/lcdm/lcdm_2_pk.dat')
lcl2=np.loadtxt('../N1_nearika/output/lcdm/lcdm_2_cl.dat')
```

```python
fig, ax = plt.subplots(figsize=(9,7))

#N=1 vs lcdm----------------------------------------------------------------------------
#ax.loglog(pk[:,0],pk[:,1],'--y',label=r'$w(N=1)_1$',alpha=1,linewidth=1)
ax.loglog(lpk[:,0],lpk[:,1],'k',label=r'$\Lambda_{CDM}1$',alpha=1,linewidth=1)

ax.loglog(pk2[:,0],pk2[:,1],'--g',label=r'$w(N=1)2$',alpha=1,linewidth=1)
ax.loglog(lpk2[:,0],lpk2[:,1],'b',label=r'$\Lambda_{CDM}2$',alpha=1,linewidth=1)




#--------------------------------------------------------------------------------------
ax.legend(loc='best', fancybox=True, framealpha=1)
ax.set_ylabel(r'$P(k) (Mpc/h)^3$',fontsize=15)
ax.set_xlabel(r'$k (h/Mpc) $',fontsize=15)
#ax.set_ylim(bottom=10**-5)
ax.set_xlim(left=2*10**-5,right=1)
ax.grid(which='major',ls=":", c='black',alpha=0.4)
#ax.set_title(r'CPL vs w(N=1) for parameters 1');
plt.savefig('power_N1_vs_lcdm.png')
```

```python
fig, ax = plt.subplots(figsize=(9,7))

#ax.semilogx(cl[:,0],cl[:,1],'--y',label=r'$w(N=1)_1$',alpha=1,linewidth=1)
ax.semilogx(lcl[:,0],lcl[:,1],'k',label=r'$\Lambda_{CDM}1$',alpha=1,linewidth=1)

ax.semilogx(cl2[:,0],cl2[:,1],'--g',label=r'$w(N=1)_2$',alpha=1,linewidth=1)
ax.semilogx(lcl2[:,0],lcl2[:,1],'b',label=r'$\Lambda_{CDM}2$',alpha=1,linewidth=1)

#ax.set_xlim(left=2, right=3000)
ax.set_xlabel(r'multipole moment $\ell$',fontsize=15)
ax.set_ylabel(r'$D_{\ell}^{TT} $',fontsize=15)
ax.legend(loc='upper right', fancybox=True, framealpha=1)
ax.grid(which='both',ls=":", c='black', alpha=0.4);

#CAMBIAR NOMBRE AQUIIIIIIIIIIIIIII

plt.savefig('tt_N1_lcdm.pdf')
```
