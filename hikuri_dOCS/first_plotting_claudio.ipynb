```python
fig, ax = plt.subplots(figsize=(13,10))

#--------------------------------------------------------------------------------------
# CPL
ax.loglog(cplpk_1[:,0],cplpk_1[:,1],label='cpl params1',alpha=0.3,linewidth=1)
ax.loglog(cplpk_2[:,0],cplpk_2[:,1],label='cpl params2',alpha=0.3,linewidth=1)

# WN=1
ax.loglog(wnpk_1[:,0],wnpk_1[:,1],label='w1 params 1',alpha=0.3,linewidth=1)
ax.loglog(wnpk_2[:,0],wnpk_2[:,1],label='w1 params 2',alpha=0.3,linewidth=1)
ax.loglog(wnpk_9[:,0],wnpk_9[:,1],label='w1 params 9',alpha=0.3,linewidth=1)

ax.loglog(cplpk_none[:,0],cplpk_none[:,1],label='cpl with no params',alpha=0.3,linewidth=1)

#--------------------------------------------------------------------------------------
ax.legend(loc='upper right', fancybox=True, framealpha=1)
ax.set_ylabel(r'$P(k) (Mpc/h)^3$',fontsize=15)
ax.set_xlabel(r'$k (h/Mpc) $',fontsize=15)
ax.set_ylim(bottom=10**-5)
ax.set_xlim(left=10**-3,right=10**3)
ax.grid(which='major',ls=":", c='black')
#ax.set_title(r'CPL vs w(N=1) for parameters 1');
```
